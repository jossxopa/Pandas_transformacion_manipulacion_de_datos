## Cambiar el tipo de dato

El dataframe tiene datos tanto strings como númerico por lo que es necesario darle el valor de tipo de dato correcto, ya que si lo dejamos así no podremos manipular los datos.

Object significa: Para python es un "no sé que tipo de dato es o es un string".

* 1. Importar Numpy: cuando trabajamos con número es necesario usar Numpy
* 2. Asegurarnos que los datos se guarden, por lo que debemos guardarlo directamente en la columna.
* 3. Analicemos nuestros datos para ahorrar tiempo y líneas podemos guardar en una variable los datos que pertenecen al mismo tipo de valor y con ello ejecutar una línea que solicite el cambio de valor.
* 4. Hay columnas que tienen un dato float y un caracter, en este caso el signo de dolar, por lo que no se podrá cambiar con el método .astype(np.float64)

```python
import numpy as np
```

```python
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)

# el primer datos['max_hospedes'] es para que ya se guarde directamente en la columna que queremos cambiar el tipo de dato
# de lo contrario no se guardara.
```

```python
# Verificamos que los datos cambiaros

datos.info()
```

```python
# Hacerlo en conjunto
columnas_numericas_int = ["cantidad_baños","cantidad_cuartos","cantidad_camas"]

datos[columnas_numericas_int] = datos[columnas_numericas_int].astype(np.int64)
# Esta funciona como una "iteración" es decir que asignara el valor int a cada elemento de la lista, ese orden.
# Solo se cambiaron valores de las columnas que tenían que cambiar a valor entero
```

```python
# Verificamos el cambio
datos.info()
```

```python
# Cambio a float

datos["evaluacion_general"] = datos["evaluacion_general"].astype(np.float64)
```

```python
datos.info()
```

### Covertir datos númericos y strings (en cojunto)

```python
datos['precio'] = datos['precio'].apply(lambda x: x.replace("$","").replace(",","").strip()) #esto solo quita el string, aún no hay cambio de tipo de valor

# Esta línea dice: de mi columna precio, aplica lo siguiente, por cada x (cada valor de lo que hay en precio) reemplaza "$" por "" (nada), y reemplaza "," por "" (nada). Tambien elimina espacios (strip)
# Nota, podemos usar los replace que queramos en una misma fila.

datos['precio'] = datos["precio"].astype(np.float64) # con esto ya se guarda como float
```

```python
datos.info()
```

### Multiples cambios en uno solo

El método apply solo funciona para series (para un sola linea de datos) por lo que no se podría utilizar con más de dos columnas.

Apply es para series
y el método para dataframes (más de dos columnas): **applymap**

```python
datos[["cuota_deposito","cuota_limpieza"]] = datos[["cuota_deposito","cuota_limpieza"]].applymap(lambda x: x.replace("$","").replace(",","").strip()) #esto solo quita el string, aún no hay cambio de tipo de valor
```

```python
# Guardamos los datos en las columnas y verificamos que ya se haya cambiado a float.
datos[["cuota_deposito","cuota_limpieza"]] = datos[["cuota_deposito","cuota_limpieza"]].astype(np.float64)
datos.info()
```

## Datos de texto

Podemos extraer palabras que nos pueda dar un valor, ejemplo "Maravilloso", "Horrible" a eso se le llama tokenización. Vamos a separar el texto por palabras.

Que palabras ayudan a la valorización del proyecto.
Usaremos el método **regex** para limpiar datos, eliminando caracteres para una correcta extracción de información y esta pueda ser valorizada.

```python
# Paso 1: convertir los datos de texto en strings y en minúscula para poder trabajar.

datos["descripcion_local"] = datos["descripcion_local"].str.lower() #str le decimos a python que es de tipo string y lower que lo pase a minúscula.
datos.head() #observamos que el texto ya esta en minúscula
```

#### Eliminar caracteres especiales con regex:

El objetivo es eliminar caracteres especiales que no aportan valor para el análisis de datos, como comas, puntos, signos de interrogación, etc.

Se utilizará **str.replace** junto con regex para eliminar caracteres no deseados

Aquí, se utiliza **str.replace** junto con una expresión regular para eliminar todos los caracteres que no sean letras (mayúsculas y minúsculas), números, guiones o comillas simples.

* **[^a-zA-Z0-9\-\']**: Esta es la expresión regular.
* El **^** dentro de los corchetes significa "todo lo que no sea". a-zA-Z representa todas las letras mayúsculas y minúsculas, 0-9 representa todos los números, \- representa un guión (debe estar dentro de una barra invertida) y \' representa una comilla simple (también escapada).
* **' '**: Este es el carácter de reemplazo, en este caso un espacio en blanco.
* **regex=True:** Este parámetro indica que se está utilizando una expresión regular.

####  Eliminar guiones aislados con regex:

'(\W)-(\W)'

**(\W)-(\W):** Esta es la expresión regular.
* **\W ** representa cualquier carácter que no sea una letra, número o guión bajo. Los paréntesis se utilizan para crear grupos de captura, pero en este caso no se utilizan para nada más. En otras palabras, si la pabrala no tiene otra palabra pegada a su derecha o izquiera no borra el guion, pero en caso de que no exista alguna a lado, sí la eliminará

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace("[^a-zA-Z0-9-\']"," ",regex = True)
datos['descripcion_local'] = datos['descripcion_local'].str.replace("(\W)-(\W)", " ",regex = True)

datos['descripcion_local'][3169]
#Observamos que elimino caracteres.
```

```python
datos.head()
```

### Tokenizacion de strings

El proceso de tokenización implica dividir una frase o texto en unidades más pequeñas, conocidas como tokens, que generalmente son palabras individuales.
Esto es útil para el análisis de texto y el machine learning, donde cada palabra puede ser tratada como una característica.

Utilizamos el método **split()** para separar frases en palabras, aprovechando el espacio como delimitador natural.
Por defecto, split() usa el espacio en blanco como separador si no se especifica otro carácter (por ejemplo una ",")



```python
# Vamos a tokenizar la columna descripcion_local

datos['descripcion_local'].str.split() #vemos que ya no estan separada por espacios, sino por coma.

# Aplica la función split() a cada elemento de la columna, dividiendo cada frase en una lista de palabras donde el separador es el espacio en blanco (por defecto).
```

```python
# Guardamos la columna

datos['descripcion_local'] = datos['descripcion_local'].str.split()
```

```python
# Ahora manipulemos datos comunidades
# 1. Limpiamos los datos ya que tiene llaves (el texto no esta almacenado en un diccionario)

datos['comodidades'].str.replace(r'\{|\"|\}', " ", regex=True)
```

```python
#  Tokenización de la columna comodidades después de la limpieza

datos['comodidades'] = datos['comodidades'].str.split(",") # Aquí si ponemos una coma dentro de los parentesis porque ese es el divisor de las palabras
```

### Extraer información de la fecha

Usamos el método strftime para personalizar el formato de la fecha y obtener solo el año y el mes.

Solo de nuesto dataframe dt_datetime queremos extraer el año y el mes para que después analicemos cuantos lugares teniamos disponibles durante el mes y así lograr identificar que meses tenemos más o menos afluencia.

```python
# Extraer el año y el mes de la columna 'fecha'

dt_datetime['fecha'].dt.strftime("%Y-%m")
```

```python
# Agrupar los datos por año y mes y sumar la cantidad de lugares disponibles

subset = dt_datetime.groupby(dt_datetime['fecha'].dt.strftime("%Y-%m"))["lugar_disponible"].sum()
subset

# El sum solo suma los valores booleanos true, que significa en nuestro dataframe que estaba disponible el lugar.
# Agrupamos primero por la fecha y despues por la columna booleana, que nos informa sobre la disponibilidad y esto se guarda en la variable subconjunto
```
